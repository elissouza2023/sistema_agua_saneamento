# 2: Carregando o arquivo base via GITHUB

```python
import pandas as pd

url = "https://raw.githubusercontent.com/elissouza2023/sistema_agua_saneamento/main/data/dados_micromedicao.xls"

# Ler diretamente do GitHub
df = pd.read_excel(url)

print("Planilha carregada com sucesso!")

# Visualizar primeiras linhas
df.head()
```

# 3: Conhecendo a base

```python
# Informações da base
df.info()
```

```python
df.describe()
```

```python
#Rótulos de colunas
df.columns.tolist()
```

```python
#Padronização das colunas, todas com letras minusculas, sem espaços, substituido Ç por c e Ã para A
df.columns = (
    df.columns
    .str.strip()
    .str.upper()
    .str.replace(' ', '_')
    .str.replace('Ç', 'C')
    .str.replace('Ã', 'A')
)
```

```python
# Verificando colunas vazias
df.isnull().sum()
```

# 4: Salvando o DataFrame

```python
# Exportando a base tratada para csv

df.to_csv(
    'dados_tratados.csv',
    index=False,
    encoding='utf-8'
)

print("CSV salvo com sucesso!")
```
