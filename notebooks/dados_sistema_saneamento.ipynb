```python
import pandas as pd

url = "https://raw.githubusercontent.com/elissouza2023/sistema_agua_saneamento/main/data/dados_micromedicao.xls"

# Ler diretamente do GitHub
df = pd.read_excel(url)

print("Planilha carregada com sucesso!")

# Visualizar primeiras linhas
df.head()
```

```python
#Padronização das colunas, todas com letras minusculas, sem espaços, substituido Ç por c e Ã para A

df.columns = (
    df.columns
    .str.strip()
    .str.upper()
    .str.replace(' ', '_')
    .str.replace('.', '')
    .str.replace('/', '_')
    .str.replace('Ç', 'C')
    .str.replace('Ã', 'A')
    .str.replace('Á', 'A')
)
```

# 4: Tratando os dados no CSV

```python
# REMOVENDO LINHAS SEM SIT_LIG_AGUA

df = df[df['SIT_LIG_AGUA'].notna()]
```

```python
# PREENCHER NULOS NUMÉRICOS COM ZERO


colunas_numericas = df.select_dtypes(
    include=['float64', 'int64']
).columns

df[colunas_numericas] = df[colunas_numericas].fillna(0)

# Visualizar primeiras linhas
df.head()
```

```python
# PREENCHIMENTO DA CATEGORIA_PRINCIPAL


df['CATEGORIA_PRINCIPAL'] = (
    df['CATEGORIA_PRINCIPAL']
    .fillna('NAO INFORMADO')
)

# Visualizar primeiras linhas
df.head(20)
```

```python
# Convertendo data
df['DATA_INSTALACAO_HIDROMETRO'] = pd.to_datetime(
    df['DATA_INSTALACAO_HIDROMETRO'],
    errors='coerce'
)
```

# Engenharia de atributos

# 6. Consumo Acumulado do período

```python
# Calculando a receita anual
colunas_consumo = [
    col for col in df.columns
    if 'VOLUME_FATURADO_' in col
]

df['CONSUMO_ACUMULADO'] = (
    df[colunas_consumo]
    .sum(axis=1)
)
```

# 7. Receita Acumulada

```python
# RECEITA ACUMULADA do período

colunas_receita = [
    col for col in df.columns
    if 'VALOR_TOTAL_' in col
]

df['RECEITA_ACUMULADA'] = (
    df[colunas_receita]
    .sum(axis=1)
)
```

# 8. Consumo médio Mensal

```python
# CONSUMO MEDIO


df['CONSUMO_MEDIO'] = (
    df[colunas_consumo]
    .mean(axis=1)
)
```

# 9. Receita média mensal

```python
# RECEITA MEDIA


df['RECEITA_MEDIA'] = (
    df[colunas_receita]
    .mean(axis=1)
)
```

# 10. Idade do Hidrometro

```python
# IDADE DO HIDROMETRO


df['IDADE_HIDROMETRO'] = (
    (
        pd.Timestamp.today() -
        df['DATA_INSTALACAO_HIDROMETRO']
    ).dt.days / 365
).round(1)
```

# 11. Ticket médio por m³

```python
# TICKET MEDIO POR M³
df['TICKET_MEDIO_M3'] = np.where(
    df['CONSUMO_ACUMULADO'] > 0,
    df['RECEITA_ACUMULADA'] / df['CONSUMO_ACUMULADO'],
    0
)
```

# 13. Indicadores Estratégicos

Tarifa média da base

```python
# TARIFA MEDIA GLOBAL

tarifa_media = (
    df['RECEITA_ACUMULADA'].sum() /
    df['CONSUMO_ACUMULADO'].sum()
)

print(f'Tarifa média da base: R$ {tarifa_media:.2f}/m³')
```

Receita potencial

```python
# RECEITA POTENCIAL

df['RECEITA_POTENCIAL'] = (
    df['CONSUMO_ACUMULADO'] *
    tarifa_media
)
```

Perda estimada

```python
# PERDA ESTIMADA

df['PERDA_ESTIMADA'] = (
    df['RECEITA_POTENCIAL'] -
    df['RECEITA_ACUMULADA']
)
```

flag suspeita

```python
# um marcador automático que identifica registros com comportamento fora do padrão esperado

# FLAG DE SUSPEITA

df['FLAG_SUSPEITA'] = np.where(
    (
        df['CONSUMO_ACUMULADO'] >
        df['CONSUMO_ACUMULADO'].quantile(0.75)
    ) &
    (
        df['TICKET_MEDIO_M3'] <
        df['TICKET_MEDIO_M3'].quantile(0.25)
    ),
    'SUSPEITO',
    'NORMAL'
)

df['FLAG_SUSPEITA'].value_counts()
```

Score de prioridade financeira

```python
# SCORE DE PRIORIDADE PARA IDENTIFICAR MAIOR PERDA E HIDROMETROS MAIS ANTIGOS PARA TROCA, FISCALIZAÇÃO E INSPEÇÃO

df['SCORE_PRIORIDADE'] = (
    (
        df['PERDA_ESTIMADA']
        .clip(lower=0)
    ) *
    (
        df['IDADE_HIDROMETRO']
        .fillna(0)
    )
)
```

Ranking financeiro

```python
top_perdas = df.nlargest(
    20,
    'PERDA_ESTIMADA'
)[
    [
        'MATRICULA',
        'TIPO_HIDROMETRO',
        'IDADE_HIDROMETRO',
        'PERDA_ESTIMADA',
        'FLAG_SUSPEITA'
    ]
]

top_perdas
```

## Faixa operacional de vida do parque dos hidrometros

```python
# CLASSIFICACAO OPERACIONAL DOS HIDROMETROS

df['STATUS_HIDROMETRO'] = np.where(
    df['IDADE_HIDROMETRO'] < 5,
    'DENTRO DA VIDA UTIL',
    np.where(
        df['IDADE_HIDROMETRO'] < 7,
        'PROXIMO DO VENCIMENTO',
        'SUBSTITUICAO RECOMENDADA'
    )
)

# Para verificar a distribuição
df['STATUS_HIDROMETRO'].value_counts()
```

Visualização Gáfica do PArque de hidrometros

```python
status_hidrometro = (
    df['STATUS_HIDROMETRO']
    .value_counts()
    .reset_index()
)

status_hidrometro.columns = [
    'STATUS',
    'QUANTIDADE'
]

fig = px.bar(
    status_hidrometro,
    x='STATUS',
    y='QUANTIDADE',
    color='STATUS',
    title='Situação Operacional dos Hidrômetros',
    text='QUANTIDADE',
    color_discrete_map={
        'DENTRO DA VIDA UTIL': 'green',
        'PROXIMO DO VENCIMENTO': 'yellow',
        'SUBSTITUICAO RECOMENDADA': 'red'
    }
)

fig.update_traces(
    textposition='outside'
)

fig.update_layout(
    xaxis_title='Status',
    yaxis_title='Quantidade de Hidrômetros',
    showlegend=False
)

fig.show()
```

STATUS_HIDROMETRO x PERDA_ESTIMADA

```python
# agrupando os dados
dados_linha = (
    df.groupby(
        ['IDADE_HIDROMETRO', 'STATUS_HIDROMETRO']
    )['PERDA_ESTIMADA']
    .mean()
    .reset_index()
)

# criando o comparativo visual
fig = px.line(
    dados_linha,
    x='IDADE_HIDROMETRO',
    y='PERDA_ESTIMADA',
    color='STATUS_HIDROMETRO',
    markers=True,
    title='Perda Estimada x Vida Útil dos Hidrômetros',
    color_discrete_map={
        'DENTRO DA VIDA UTIL': 'green',
        'PROXIMO DO VENCIMENTO': 'yellow',
        'SUBSTITUICAO RECOMENDADA': 'red'
    }
)

fig.update_layout(
    xaxis_title='Idade do Hidrômetro (anos)',
    yaxis_title='Perda Estimada Média (R$)',
    template='plotly_white'
)

fig.show()
```
